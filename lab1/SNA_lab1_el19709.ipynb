## Α. Κατασκευή και οπτικοποίηση σύνθετων τύπων δικτύου

### Τοπολογία REG για βαθμό πλέγματος $d = 4$

```python
random.seed(seedno)
n=49 # μέγεθος δικτύου: αντικαταστήστε το x με το τελευταίο ψηφίο του αριθμού μητρώου σας.
d=4
# για την κατασκευή του REG χρησιμοποιήστε την watts_strogatz_graph από το networkx.
# Επιλέξτε προσεκτικά την τιμή της παραμέτρου p.
REG=nx.watts_strogatz_graph(n, d, 0, seed=seedno)
#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(REG)
```

### Τοπολογία RG (ER) για αριθμό συνδέσεων $M = 150$

```python
random.seed(seedno)
n=49
M = 150

# για την κατασκευή του RGER χρησιμοποιήστε την gnm_random_graph από το networkx
RGER= nx.gnm_random_graph(n,M,seed=seedno,directed=False)

# οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(RGER)
```

### Τοπολογία RGG  με ακτίνα σύνδεσης $R=0.35$

```python
random.seed(seedno)
n=49
R=0.35

# για την κατασκευή του RGG χρησιμοποιήστε την random_geometric_graph από το networkx
RGG=nx.random_geometric_graph(n,R,dim=2,seed=seedno)

#οπτικοποίηση: https://networkx.org/documentation/stable/auto_examples/drawing/plot_random_geometric_graph.html
# position is stored as node attribute data for random_geometric_graph
pos = nx.get_node_attributes(RGG, "pos")

# find node near center (0.5,0.5)
dmin = 1
ncenter = 0
for n in pos:
    x, y = pos[n]
    d = (x - 0.5) ** 2 + (y - 0.5) ** 2
    if d < dmin:
        ncenter = n
        dmin = d

# color by path length from node near center
p = dict(nx.single_source_shortest_path_length(RGG, ncenter))

plt.figure(figsize=(8, 8))
nx.draw_networkx_edges(RGG, pos, alpha=0.4)
nx.draw_networkx_nodes(
    RGG,
    pos,
    nodelist=list(p.keys()),
    node_size=80,
    node_color=list(p.values()),
    cmap=plt.cm.Reds_r,
)

plt.xlim(-0.05, 1.05)
plt.ylim(-0.05, 1.05)
plt.axis("off")
plt.show()
```

### Τοπολογία SF με παράμετρο επισύναψης $d=4$

```python
random.seed(seedno)
n=49
d=4

# για την κατασκευή του SF χρησιμοποιήστε την barabasi_albert_graph από το networkx
SF= nx.barabasi_albert_graph(n,d,seed=seedno)

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(SF)
```

### Τοπολογία SW με βαθμό αρχικού πλέγματος $d=4$ και πιθανότητα ανασύνδεσης $p=0.25$

```python
random.seed(seedno)
n=49
d=4
p=0.25

# για την κατασκευή του SW χρησιμοποιήστε την watts_strogatz_graph από το networkx.
SW=nx.watts_strogatz_graph(n,d,p,seed=seedno)

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(SW)
```

## B. Μετρικές δικτύου

### B1. Συντελεστής ομαδοποίησης (Clustering coefficient)

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β1.1 Να υπολογιστεί ο συντελεστής ομαδοποίησης κάθε κόμβου και να αναπαρασταθεί η κατανομή του.
#### Β1.2  Να υπολογιστεί ο μέσος συντελεστής ομαδοποίησης και η διασπορά του.
#### Β.1.3 Να βρεθούν οι κόμβοι με το μικρότερο και το μεγαλύτερο συντελεστή ομαδοποίησης.
#### Να σχολιαστούν τα αποτελέσματα.

```python
import operator
def cc(G):
    x=nx.clustering(G)
    #print(x)
    lcc=list(x.values());
    n,bins,patches=plt.hist(lcc,range=(0,1),bins=100)
    plt.show()
    mo=np.mean(lcc)
    var=np.var(lcc)
    print("Mean cc: ",mo)
    print("Variance of cc: ",var)
    maxcc=max(x, key=x.get)
    mincc=min(x, key=x.get)
    print("Node with max cc: ",maxcc)
    print("Node with min cc: ",mincc,"\n")

print("REG:")
cc(REG)
print("RGER:")
cc(RGER)
print("RGG:")
cc(RGG)
print("SF:")
cc(SF)
print("SW:")
cc(SW)
```

### REG
Ο συντελεστής ομαδοποίησης είναι ο ίδιος για όλους τους κόμβους, γεγονός που αναμέναμε λόγω της ομοιομορφίας των regular γραφημάτων.

### RGER
Στον τυχαίο αυτό γράφο η πιθανότητα να υπάρχει ακμή μεταξύ 2 κόμβων είναι 150/(49*48/2)~=0.128, άρα η πιθανότητα δύο γειτώνων ενός κόμβου να είναι επίσης γείτονες μεταξύ τους είναι περίπου αυτή. Άρα οι χαμηλές τιμές του συντελεστή ομαδοποίησης και η μέση τιμ΄΄η που ισούται με περίπου 0.1 δικαιολογούνται.

### RGG
Σε αυτήν την τοπολογία παρατηρούμε τις ψηλότερες τιμές του συντελεστή ομαδοποίησης, γεγονός που εξηγείται αν σκεφτεί κανείς ότι 2 γείτονες ενός κόμβου (δηλαδή κόμβοι που βρίσκονται σε μικρότερη απόσταση ρ από αυτον) είναι αρκετά πιθανό να βρίσκονται και σε μικρότερη απόσταση ρ μεταξύ τους.

### SF
Παρατηρούμε παρόμοια αποτελέσματα όπως με την τοπολογία RGER λόγω τυχαιότητας.

### SW
Πάλι λόγω τυχαιότητας οι τιμές του συντελεστή είναι χαμηλές (και σχεδόν όλες χαμηλότερες από τη regular τοπολογία).

### Β2. Μέσο μήκος μονοπατιού - Διάμετρος - Ακτίνα γράφου

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β2.1 Nα υπολογιστεί το μέσο μήκος μονοπατιού και η διασπορά του. Προτείνεται η χρήση της συνάρτησης  all_pairs_shortest_path_length του networkx.
#### B2.2 Να υπολογιστεί η διάμετρος (συνάρτηση diameter) και η ακτίνα (συνάρτηση radius).
#### Για τα παραπάνω πραγματοποιήστε συγκρίσεις και εξηγήστε με βάση τα χαρακτηριστικά της κάθε τοπολογίας όπως παρουσιάζονται στη θεωρία.

```python
def compute_path_lengths(G):
    apsp=nx.all_pairs_shortest_path_length(G)
    lapsp=list(apsp)
    #print(lapsp)
    lop=[]
    totn=G.number_of_nodes()
    for node in lapsp:
        n1=node[0]
        lte=list(node[1].values())
        for i in range(n1+1,totn):
            if i in node[1].keys():
                lop.append(node[1][i])
    mo=np.mean(lop)
    va=np.var(lop)
    print("Mean shortest path length: ",mo)
    print("Variance of shortest path length: ",va)
    diam=nx.diameter(G)
    rad=nx.radius(G)
    print("Diameter of graph: ",diam)
    print("Radius of graph: ",rad,"\n")

print("REG:")
compute_path_lengths(REG)
print("RGER:")
compute_path_lengths(RGER)
print("RGG:")
compute_path_lengths(RGG)
print("SF:")
compute_path_lengths(SF)
print("SW:")
compute_path_lengths(SW)
```

###REG
Λόγω της ομοιομορφίας του γραφήματος το μονοπάτι μεταξύ 2 κόμβων μπορεί να έχει μήκος από 1 μέχρι και 12, και η κατανομή της απόστασης είναι σχεδόν ομοιόμορφη, και γιαυτό η μέση τιμή 6,5 και η διασπορά 11.91 είναι λογικά αποτελέσματα. Σημειώνεται επίσης πως είναι οι μεγαλύτερες μετρικές για τις τοπολογίες που παρουσιάζονται εδώ. Η διάμετρος και η ακτίνα είναι ίσες πάλι λόγω ομοιομορφίας

###RGER
Λόγω της τυχαιότητας του γραφήματος καταλαβαίνουμε ότι τα μονοπάτια μεταξύ 2 κόμβων θα έχουν παραπλήσιο μήκος, αλλά και σχετικά μικρό. Αυτό εξηγέι τις χαμηλές τιμές της μέσης τιμής και της διασποράς. Η διάμετρος έχει τιμή 5, ενώ η ακτίνα 3.

###RGG
Πάλι το μέσο μήκος μονοπατιού είναι μικρό αφού συνήθως δεν θα χρειαστεί να κάνεις πολλά βήματα για να πας από έναν κόμβο σε άλλο. Οι περιπτώσεις που ανεβάζουν την μέση τιμή είναι όταν διαλέγουμε ζευγάρι κόμβων που βρίσκονται σε απέναντι άκρα του χώρου μας. Όμως και πάλι αυτές δεν είναι πολλές αφού το μέσο μήκος είναι αρκετά μικρό, και η διασπορά επίσης. Οι τιμές της διαμέτρου και της ακτίνας είναι ίδιες με την τοπολογία RGER. Εδώ σημειώνεται ότι σε μργαλύτερα γραφήματα με περισσότερους κόμβους το πιθανότερο το μέσο μήκος μονοπατιού να ήταν αρκετά μεγαλύτερο από τις άλλες τοπολογίες.

###SF
Λόγω του τρόπου κατασκευής του γραφήματος (κάθε κόμβος που προστίθεται ενώνεται με μεγαλύτερη πιθανότητα σε αυτό με μεγάλο βαθμό ακμών), οι χαμηλές τιμές των αποτελεσμάτων είναι αναμενόμενες. Χαμηλή μέση τιμή και διασπορά, αλλά επίσης και διάμετρος και ακτίνα - Οι ελάχιστες από όλες τις τοπολογίες που παρουσιάζουμε εδώ.

###SW
Το μέσο μήκος είναι σχετικά μικρό, και η διασπορά το ίδιο. Αυτό εξηγείται επειδή υπάρχουν κόμβοι με μεγάλο βαθμό, ενώ άλλοι με μικρό, και έτσι αυτοί με μεγάλο μπορούν να είναι οι ενδιάμεσοι κόμβοι στα μονοπάτια. Η διάμετρος και η ακτίνα είναι αναμενόμενα μεγαλύτερες από αυτές της SF τοπολογίας.

#### B.2.3 Για κάθε τοπολογία, ποιος κόμβος συνδέεται με τους περισσότερους κόμβους με μήκος ελάχιστου μονοπατιού ίσο με τη διάμετρο της τοπολογίας?

```python
def find_nodes(G):
    diam=nx.diameter(G)
    apsp=nx.all_pairs_shortest_path_length(G)
    lapsp=list(apsp)

    maxt=0;
    nans=-1;
    ans=[];
    for node in lapsp:
        n1=node[0]
        lte=list(node[1].values())
        temp=[]
        for i in range(0,len(node[1])):
            if i in node[1].keys():
                if(node[1][i]==diam):
                    temp.append(i)
        if(len(temp)>maxt):
            maxt=len(temp)
            ans=temp.copy()
            nans=n1
    print("Node that has shortest path equal with the diameter with the most number other nodes: ",nans)
    print("Nodes that it connects with: ",ans,"\n")
```

```python
# να προσδιοριστεί το label του κόμβου και το σύνολο των κόμβων
# με τους οποίους συνδέεται με μήκος ελάχιστου μονοπατιού ΄ίσο με τη διάμετρο.
print("REG:")
find_nodes(REG)
print("RGER:")
find_nodes(RGER)
print("RGG:")
find_nodes(RGG)
print("SF:")
find_nodes(SF)
print("SW:")
find_nodes(SW)
```

#### Περιφέρεια του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με τη διάμετρό του.
#### Κέντρο του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με την ακτίνα του.  
#### Β.2.4 Για κάθε τοπολογία, να βρεθεί η περιφέρεια και το κέντρο της και να σχολιαστούν τα αποτελέσματα.

```python
def find_periphery(G):
    diam=nx.diameter(G)
    x=nx.eccentricity(G)
    per=[]
    for i in range(0,G.number_of_nodes()):
        if x[i]==diam:
          per.append(i)
    print("Periphery of graph: ",per)
```

```python
def find_center(G):
    rad=nx.radius(G)
    x=nx.eccentricity(G)
    cen=[]
    for i in range(0,G.number_of_nodes()):
        if x[i]==rad:
          cen.append(i)
    print("Center of graph: ",cen,"\n")

print("REG:")
find_periphery(REG)
find_center(REG)
print("RGER:")
find_periphery(RGER)
find_center(RGER)
print("RGG:")
find_periphery(RGG)
find_center(RGG)
print("SF:")
find_periphery(SF)
find_center(SF)
print("SW:")
find_periphery(SW)
find_center(SW)
```

###REG

###RGER
Παρατηρούμε ότι λίγοι κόμβοι ανήκουν στην περιφέρεια του γράφου, και ότι περισσότεροι ανήκουν στο κέντρο. Αυτό σημαίνει ότι λίγοι κόμβοι βρίσκονται μακριά μεταξύ τους.

###RGG
Παρατηρούμε όμοια αποτελέσματα με την τοπολογία RGER.

###SF
Αφού η διάμετρος και η ακτίνα διαφέρουν μόνο κατά 1, όλοι οι κόμβοι θα ανήκουν είτε στην περιφέρεια έιτε στο κέντρο. Εδώ οι περισσότεροι ανήκουν στην περιφέρεια, γεγονός που σημαίνει ότι απέχουν τουλάχιστον απο έναν άλλο κόμβο απόσταση 3.

###SW
Μόνο 2 κόμβοι ανήκουν στην περιφέρεια, άρα ουσιαστικά αυτοί οι 2 είναι τα άκρα της διαμέτρου. Στο κεντρο ανήκουν 3 κόμβοι. Άρα οι υπόλοιποι απέχουν απόσταση 4 από τον μακρυνότερό τους κόμβο.

## Γ) Mετρικές κεντρικότητας κόμβων

### Γ1. Κεντρικότητα βαθμού (Degree centrality)
##### Γ1.1 Να αναπαρασταθεί η κατατανομή βαθμών κόμβων για κάθε τοπολογία.
##### Γ1.2 Να υπολογιστεί ο μέσος βαθμός κόμβου καθώς και η διασπορά των βαθμών κόμβου για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_degree(G):
    deg=[]
    for i in range(0,G.number_of_nodes()):
        deg.append(G.degree(i))
    plt.hist(deg)
    plt.show()
    mo=np.mean(deg)
    va=np.var(deg)
    print("Mean node degree: ",mo)
    print("Variance of node degree: ",va,"\n")

print("REG:")
find_degree(REG)
print("RGER:")
find_degree(RGER)
print("RGG:")
find_degree(RGG)
print("SF:")
find_degree(SF)
print("SW:")
find_degree(SW)
```

###REG
H κατανομή είναι συνάρτηση dirac στην τιμή 4.

###RGER
Τείνει να μοιάσει με κατανομή Poisson.

###RGG
Δεν βλέπουμε την κατανομή να έχει κάποια κορυφή, κάπου, οπότε θα μπρούσαμε να πούμε ότι τείνει στην ομοιόμορφη κατανομή.

###SF
Μάλλον η πιο ξεκάθαρη (εκτός από την regular τοπολογία), η κατανομή είναι power-law.

###SW
Μοιάζει με κανονική κατανομή, αλλά στην πραγματικότητα είναι heavy-tail.

### Γ2. Κεντρικότητα εγγύτητας (Closeness centrality)
##### Γ2.1 Να αναπαρασταθεί η κατατανομή της εγγύτητας κόμβων για κάθε τοπολογία.
##### Γ2.2 Να υπολογιστεί η μέση εγγύτητα καθώς και η διασπορά της για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_closeness(G):
    x=nx.closeness_centrality(G)
    lcen=list(x.values())
    plt.hist(lcen,range=(0,1),bins=100)
    plt.show()
    mo=np.mean(lcen)
    va=np.var(lcen)
    print("Mean node closeness: ",mo)
    print("Variance of node closeness: ",va,"\n")

print("REG:")
find_closeness(REG)
print("RGER:")
find_closeness(RGER)
print("RGG:")
find_closeness(RGG)
print("SF:")
find_closeness(SF)
print("SW:")
find_closeness(SW)
```

###REG
Η κεντρικότητα είναι χαμηλή σχετικά με τις υπόλοιπες τοπολογίες, αφού γενικά οι κόμβοι απέχουν αρκετά περισσότερο από τους υπόλοιπους (μέσο μήκος μονοπατιού = 6.5). Η μηδενική διασπορά οφείλεται στη συμμετρία του γραφήματος.

###RGER
Λόγω τυχαιότητας, οι τιμές κεντρικότητας είναι κοντά στο 0.5.

###RGG
Σε αυτή την τοπολογία εμφανίζονται συχνά μικρές κλίκες, και γιαυτό φαίνεται να παρατηρούνται λίγο μεγαλύτερες τιμές κεντρικότητας από την RGER τοπολογία.

###SF
Όπως ήταν αναμενόμενο, σε αυτή την τοπολογία παρατηρούμε τις μεγαλύτερες τιμές κεντρικότητας, αφού όλοι οι κόμβοι είναι αρκετά μαζεμένοι (διάμετρος=3).

### Γ3. Ενδιαμεσική κεντρικότητα (Betweenness centrality)
##### Γ3.1 Να αναπαρασταθεί η κατατανομή της ενδιαμεσικής κεντρικότητας κόμβων για κάθε τοπολογία.
##### Γ3.2 Να υπολογιστεί η μέση ενδιαμεσική κεντρικότητα καθώς και η διασπορά της για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_betweenness(G):
    x=nx.betweenness_centrality(G,seed=seedno)
    lbet=list(x.values())
    plt.hist(lbet,range=(0,1),bins=100)
    plt.show()
    mo=np.mean(lbet)
    va=np.var(lbet)
    print("Mean node betweenness: ",mo)
    print("Variance of node betweenness: ",va,"\n")

print("REG:")
find_betweenness(REG)
print("RGER:")
find_betweenness(RGER)
print("RGG:")
find_betweenness(RGG)
print("SF:")
find_betweenness(SF)
print("SW:")
find_betweenness(SW)
```

###REG
Πάλι παρατηρείται μηδενική διασπορά λόγω συμμετρίας. Επίσης η τιμή της ενδιαμεσικής κεντρικότητας είναι η μεγαλύτερη από όλες τις τοπολογίες, αφού όλοι οι κόμβοι ανάκουν σε ελάχιστα μονοπάτια αντίστοιχων άλλων κόμβων.

###RGER
Λόγω τυχαιότητας κάποιοι κόμβοι ανήκουν σε αρκετά ελάχιστα μονοπάτια, ενώ άλλοι σε λίγα ή κανένα. Λόγω της συνεισφοράς των δεύτερων, οι τιμές κεντρικόιτητας πέφτουν.

###RGG
Όμοια με RGER, υπάρχουν κόμβοι οι οποίοι θα ανήκουν σε αρκετά μονοπάτια (αυτοί στο κέντρο του χώρου), και κόμβοι που δεν ανήκουν σε κανένα (όπως αυτοί που βρίσκονται στα άκρα του χώρου).

###SF
Εδώ πολύ λίγοι κόμβοι έχουν σχετικά μεγάλη τιμή κεντρικότητας λόγω της φύσης της τοπολογίας, δηλαδή της κατανομής power-law του βαθμού ακμής. Οι κόμβοι με μεγάλο βαθμό, είναι αυτοί με μεγάλη κεντρικότητα, αφού συνδέουν πολλούς άλλους κόμβους, και άρα θα βρίσκονται σε πολλά ελάχιστα μονοπάτια. Οι πλείστοι έχουν πολύ χαμηλή ενδιαμεσική κεντρικότητα.

###SW
Παρουσιάζει παρόμοια συμπεριφορά με τις υπόλοιπες τοπολογίες, μόνο που τώρα οι διαφορες τιμές κεντρικότητας είναι πιο συχνές (πχ σε σχέση με την τοπολογία SF).

### Γ4. Kεντρικότητα Katz
##### Γ4.1 Να αναπαρασταθεί η κατανομή της κεντρικότητας Katz των κόμβων για κάθε τοπολογία.
Για τον προσδιορισμό της παραμέτρου $alpha$ της συνάρτησης katz_centrality,  να υπολογιστεί η μεγαλύτερη ιδιοτιμή $λ_{max}$ του πίνακα γειτνίασης με τη συνάρτηση adjacency_spectrum του networkx.
H $alpha$ ορίζεται να είναι ιση με $\frac{1}{λ_{max}}-0.01$.
##### Γ4.2 Να υπολογιστεί η μέση  κεντρικότητα Katz  καθώς και η διασπορά της για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_katz(G):
    A=nx.adjacency_matrix(G)
    A=A.todense()
    eigenvalues,eigenvec=np.linalg.eig(A)
    l=max(eigenvalues)
    a=1/l-0.01
    x=nx.katz_centrality(G,alpha=a)
    lkat=list(x.values())
    plt.hist(lkat,range=(0,1),bins=100)
    plt.show()
    mo=np.mean(lkat)
    va=np.var(lkat)
    print("Mean Katz centrality: ",mo)
    print("Variance of Katz centrality: ",va,"\n")

print("REG:")
find_katz(REG)
print("RGER:")
find_katz(RGER)
print("RGG:")
find_katz(RGG)
print("SF:")
find_katz(SF)
print("SW:")
find_katz(SW)
```

###REG
Πάλι η διασπορά είναι μηδενική λόγω συμμετρίας.

###RGER, RGG, SF, SW
Παρατηρείται παρόμοια συμπεριφορά, με μέσες τιμές κεντρικότητας Katz παραπλήσιες, και μικρότερες από την τοπολογία regular.

#### Να προσδιοριστούν, για κάθε τοπολογία, οι πιο κεντρικοί κόμβοι ως προς το βαθμό, την εγγύτητα, την ενδιαμεσική κεντρικότητα και την κεντρικότητα Katz. Να σχολιαστούν τα αποτελέσματα.

```python
def find_central(G):
    deg=nx.degree_centrality(G)
    clo=nx.closeness_centrality(G)
    bet=nx.betweenness_centrality(G)
    A=nx.adjacency_matrix(G)
    A=A.todense()
    eigenvalues,eigenvec=np.linalg.eig(A)
    l=max(eigenvalues)
    a=1/l-0.01
    kat=nx.katz_centrality(G,alpha=a)

    maxdeg={ k for k, v in deg.items() if v == max(deg.values()) }
    maxclo={ k for k, v in clo.items() if v == max(clo.values()) }
    maxbet={ k for k, v in bet.items() if v == max(bet.values()) }
    maxkat={ k for k, v in kat.items() if v == max(kat.values()) }
    print("Max degree centrality: ",max(deg.values()),"\nNode(s) with max degree centrality: ",maxdeg)
    print("Max closeness centrality: ",max(clo.values()),"\nNode(s) with max closeness centrality: ",maxclo)
    print("Max betweenness centrality: ",max(bet.values()),"\nNode(s) with max betweeness centrality: ",maxbet)
    print("Max Katz centrality: ",max(kat.values()),"\nNode(s) with max Katz centrality: ",maxkat,"\n")


print("REG:")
find_central(REG)
print("RGER:")
find_central(RGER)
print("RGG:")
find_central(RGG)
print("SF:")
find_central(SF)
print("SW:")
find_central(SW)
```

Στην τοπολογία regular όλοι οι κόμβοι είναι και οι πιο κεντρικοί, λογω συμμετρίας.

Στις υπόλοιπες παρατηρούνται το πολύ 3 κόμβοι να είναι πιο κεντρικοί ως προς ένα είδους κεντρικότητας.

Επίσης παρατηρείται, ότι αν ένας κόμβος είναι ο πιο κεντρικός ως προς ένα είδος κεντρικότητας, τότε πολύ πιθανό να είναι και ο πιο κεντρικός και ως προς άλλο είδος.

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS) που εξετάστηκαν στις ενότητες Β και Γ, να υπολογιστεί με τη συνάρτηση minimum_node_cut ο ελάχιστος αριθμός κόμβων που πρέπει να αφαιρεθούν για να προκύψει μη συνδεδεμένο δίκτυο.
#### Να σχολιαστούν τα αποτελέσματα.

```python
def find_cut(G):
    cutset=nx.minimum_node_cut(G)
    print("Minimum number of nodes to cut: ",len(cutset))

print("REG:")
find_cut(REG)
print("RGER:")
find_cut(RGER)
print("RGG:")
find_cut(RGG)
print("SF:")
find_cut(SF)
print("SW:")
find_cut(SW)
```

###REG
Για την τοπολογία regular το αποτέλεσμα ήταν αναμενόμενο αφού κάθε κόμβος συδέεται με 4 άλλους, οπότε για να απομονώσουμε έναν οποιονδήποτε πρέπει να αφαιρέσουμε 4 κόμβους.

###RGER
Λόγω τυχαιότητας, δεν υπάρχει συγκεκριμένος λόγος που χρειάζεται να αφαιρεθεί 1 μόνο κόμβος για να γίνει το γράφημα μη συνεκτικό

###RGG
Πάλι σχετικά τυχαίο αποτέλεσμα, οι κόμβοι όμως που θα αφαιρεθούν θα βρίσκονται κεντρικά στον χώρο, η μπορεί να απομονώνουν ένα μικρό τμήμα του γραφήματος στα άκρα του.

###SF
Σαν γράφημα είναι πολύ συνδεδεμένο, οπότε είναι λογικό να χρειάζεται σχετικά μεγάλος αριθμός κόμβων να αφαιρεθούν έτσι ώστε να γίνει μη συνεκτικό

###SW
Πάλι λόγωτ τυχαιότητας, δεν υπάρχει συγκεκριμένος λόγος για το αποτέλεσμα αυτό.

### E) Μελέτη της εξελικτικής μετατροπής δικτύου REG σε δίκτυο SW και RGER

##### Μεταβάλλοντας την παράμετρο $g_p$ από $0$ έως $1$ με βήμα $0.1$της συνάρτησης που κατασκευάζει τοπολογίες SW μεγέθους $n=30x$ ( αντικαταστήστε το x με το τελευταίο ψηφίο του ΑΜ) με βαθμό αρχικού πλέγματος $d=4$, να βρείτε το μέσο μήκος μονοπατιού και το μέσο συντελεστή ομαδοποίησης. Πώς συμπεριφέρεται το παραγόμενο δίκτυο για $g_p = 0$, πώς για $g_p = 1$ και πώς για ενδιάμεσες τιμές $g_p \simeq 0.1 - 0.5$;

```python
random.seed(seedno)
n=309
d=4

msp=[]
mcc=[]
xax=[]
for i in range(0,11,1):
    gp=i*0.1
    xax.append(gp)
    gr=nx.watts_strogatz_graph(n,d,gp,seed=seedno)
    x=nx.clustering(gr)
    lcc=list(x.values())
    mocc=np.mean(lcc)
    mcc.append(mocc)

    apsp=nx.all_pairs_shortest_path_length(gr)
    lapsp=list(apsp)
    #print(lapsp)
    lop=[]
    totn=gr.number_of_nodes()
    for node in lapsp:
        n1=node[0]
        lte=list(node[1].values())
        for i in range(n1+1,totn):
            if i in node[1].keys():
                lop.append(node[1][i])
    mosp=np.mean(lop)
    msp.append(mosp)
    print("Probability gp:",gp)
    nx.draw_circular(gr)
    plt.show()
    print("Mean shortest path length: ",mosp)
    print("Mean clustering coefficient: ",mocc)

plt.plot(xax,msp,'b-')
plt.title("Mean shortest path length")
plt.show()

plt.plot(xax,mcc,'b-')
plt.title("Mean clustering coefficient")
plt.show()
```

Για gp=0 το γράφημα είναι regular.

Για gp=1 το γράφημα συμπεριφέρεται τυχαία (είναι random graph).

Για 0.1<gp<0.5 το γράφημα συμπεριφέρεται ως small world.

#### Ποιοι ήταν οι 5 κεντρικότεροι χαρακτήρες του πρώτου κύκλου επεισοδίων ως προς το βαθμό, την εγγύτητα και την ενδιαμεσικότητα ?
